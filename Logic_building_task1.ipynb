```python
#1 User login Check

# stored login details
right_username = "admin"
right_password = "1234"

# get inpuit form the user
username = input("Enter username:  ")
password = input("Enter password: ")

#check the data is correct or Not

if username == right_username and password == right_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#2 pass / fail analyzer

#list of marks 
marks = [45,78,90,33,60]

#check the pass and fail
pass_count = 0 # store the number of passed student
fail_count = 0 # store the number of failed student

#check the each students marks
for m in marks:
    if m >= 50:
        pass_count +=1 
    else:
        fail_count +=1
        
print("Total passed student:",pass_count)
print("Total failed student:",fail_count)
```

```python
#3 Simple data cleaner

#user name
names = ["Alice","bob","CHARLIE"]
cleaned_names = []

#cleaned the name
for name in names:
    name = name.strip() #remove the extra spaces
    name = name.lower() #convert to the lowercase
    cleaned_names.append (name)

print("Cleaned names:", cleaned_names)
    
```

```python
#4 Message length analyzer

messages = ["Hi","Welcome to the platform","OK"]

for msg in messages:
    length = len(msg)
    
    print("Messages:",msg)
    print("Length:",length)
    
    if length > 10:
          print("This message is longer than 10 characters")
          print()
```

```python
#5 Error message detector

logs = ["INFO","ERROR","WARNING","ERROR"]

error_count = logs.count("ERROR")
print("Total error messages:",error_count)
```
